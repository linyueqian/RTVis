```python
import pandas as pd
df = pd.read_excel('demo_dataset_new.xlsx')
```

```python
df.head()
```

```python
count = df['Date'].value_counts()
```

```python
# filter date that follows the format of 'yyyy-mm-dd'
df['Date'] = df['Date'].astype(str).str.extract(r'(\d{4}-\d{2}-\d{2})')
```

```python
df['Date'].count()
```

```python
df1 = df.groupby(['Date']).size().reset_index(name='counts')
```

```python
# find the date in df, add one column based on the count of the date from df1
df['Counts'] = df['Date'].map(df1.set_index('Date')['counts']).fillna(0).astype(int)
```

```python
# drop nan and empty values
df = df.dropna()
df = df.drop(df[df['Date'] == 'nan'].index)
```

```python
# convert date to numeric
df['rank'] = pd.to_datetime(df['Date']).rank(method='dense').astype(int)
```

```python
# save the dataframe to a new excel file
df.to_excel('demo_dataset_new.xlsx', index=False)
```
